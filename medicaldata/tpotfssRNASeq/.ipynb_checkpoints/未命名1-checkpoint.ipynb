```python
import pandas as pd
import sys
import numpy as np
X_train = pd.read_csv('Xtrain.csv')
from fancyimpute import KNN, MatrixFactorization, SimpleFill, IterativeSVD
```

```python
import random
```

```python
def missdata(data,per,miss_list):
    for i in range(data.shape[0]):
        for j in range(data.shape[1]):
            if random.random()>1-per and data.loc[i][j] is not None:
                data.loc[i][j] = None
                miss_list.append([i,j])
    print('#############',len(miss_list),'#############')
    return [data,miss_list]
```

```python
def missdata_0(data0,per=0.1):
    #data = pd.read_csv(sys.argv[1])
    data = data0.copy()
    miss_list = []
    for i in range(data.shape[0]):
        for j in range(data.shape[1]):
            if random.random()>1-per:
                data.loc[i][j] = None
                miss_list.append([i,j])
    return [data,miss_list]
```

```python
def RMSE(X_train,X_miss,miss_list):
    X_train = np.array(X_train)
    X_miss = np.array(X_miss)
    loss_list = []
    for loc in miss_list:
        i,j = loc
        loss_list.append((X_train[i][j] - X_miss[i][j])*(X_train[i][j] - X_miss[i][j])/(X_train[i][j]*X_train[i][j]+0.01))
    print(np.average(loss_list))
    return np.average(loss_list)
```

```python
def RMAE(X_train,X_miss,miss_list):
    X_train = np.array(X_train)
    X_miss = np.array(X_miss)
    loss_list = []
    for loc in miss_list:
        i,j = loc
        loss_list.append(abs(X_train[i][j] - X_miss[i][j])/(abs(X_train[i][j])+0.01))
    print(np.average(loss_list))
    return np.average(loss_list)
```

```python
acc_result = []
acc_list = {}
for j in range(10):
    for i in [0.05,0.1,0.2,0.3,0.4]:
        acc_list[i] = {}
        X_train_miss, miss_list =  missdata(X_train,i)
        X_miss = SimpleFill().fit_transform(X_train_miss)
        acc_list[i]['mean'] = RMAE(X_train,X_miss,miss_list)
        X_miss = KNN().fit_transform(X_train_miss)
        acc_list[i]['KNN'] = RMAE(X_train,X_miss,miss_list)
        X_miss = IterativeSVD().fit_transform(X_train_miss)
        acc_list[i]['IterativeSVD'] = RMAE(X_train,X_miss,miss_list)
        X_miss = MatrixFactorization().fit_transform(X_train_miss)
        acc_list[i]['MatrixFactorization'] = RMAE(X_train,X_miss,miss_list)
acc_result.append(pd.DataFrame(acc_list))
```
